# Retropropagacion

```python
# Entrada y salida
x = 1
y = 2

# Peso inicial 
w = 0.2

# Tasa de aprendizaje
learning_rate = 0.3

# Entrenamos varias veces
for epoch in range(20):
    # Multiplicar el inicial contra el peso
    y_res = x * w

    # calcular el error
    error = y - y_res

    # Ajustar el peso 
    w = w + learning_rate * error * x

    # Mostrar progreso
    print(f"Época {epoch+1}: y_pred = {y_res:.4f}, error = {error:.4f}, peso = {w:.4f}")
```

## Retropropagación

La **retropropagación** es el proceso de ajustar los pesos de la red neuronal.

Primero se hace un ciclo donde la red produce una salida.  
Si la salida no es la esperada, al final del ciclo se calcula el error  
y ese error se usa para cambiar los pesos.

En el siguiente ciclo, con los pesos corregidos, la red produce una salida  
más cercana al objetivo.
