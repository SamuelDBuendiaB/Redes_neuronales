## Funciones de activación
Las funciones de activación se encargan de introducir no linealidades en
las redes neuronales, permitiendo que el modelo aprenda relaciones
complejas entre los datos. Sin ellas, la red solo podría aprender
relaciones lineales, lo cual es insuficiente para la mayoría de problemas
del mundo real.

* Lineal = Multiplicar entrada de datos -> Multiplica la salida
* No lineal = No crece de manera proporcional cambios pequeños a la entrada pueden ser grandes a la salida

```python
import numpy as np

# Entradas
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Salidas esperadas 
y = np.array([[0], [1], [1], [0]])
```

## Red Lineal

```python
# Pesos 
W = np.random.randn(2, 1)
b = np.random.randn(1)

# Salida de la "red"
output = X @ W + b

print("Salida sin activación:")
print(output)
```

## Red no lineal

```python
def relu(x):
    return np.maximum(0, x)

# Capa oculta
W1 = np.random.randn(2, 4)
b1 = np.random.randn(4)

# Capa de salida
W2 = np.random.randn(4, 1)
b2 = np.random.randn(1)

z1 = X @ W1 + b1
a1 = relu(z1)          
z2 = a1 @ W2 + b2

print("Salida con activación:")
print(z2)

```
